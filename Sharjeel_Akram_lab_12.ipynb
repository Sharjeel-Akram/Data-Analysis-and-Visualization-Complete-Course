#**Unsupervised Learning**


```python

import pandas as pd
import numpy as np
import random
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("seeds - seeds.csv")
df.head()
```

```python
remove = ['ID','seedType']
new_df = df[df.columns.difference(remove)]
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
scaled_data = scalar.fit_transform(new_df)
pd.DataFrame(scaled_data).head(2)
```

```python
from sklearn.metrics import silhouette_score

model=KMeans(n_clusters=5,random_state=1)
kmeans = model.fit(scaled_data)
cluster_preds=model.fit_predict(scaled_data)
labels = kmeans.labels_
score=silhouette_score(scaled_data,labels,metric='euclidean')
print("Silhouette_score for clusters is", score)
```
