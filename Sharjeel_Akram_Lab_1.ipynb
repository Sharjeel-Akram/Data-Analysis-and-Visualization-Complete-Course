# Hints:
As this is a transcribed version of a speech, punctuation marks such as comma, colon, semi-colon, period, hyphen are present. To ensure uniform results remove such characters and keep only those words that contain alphabets, solely. 

Same words that are present in different cases, i.e. upper case and lower case should be considered as one. For example, "Speech" is the same as "speech" and should add to the count in lowercase.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the speech.txt file and create a word list
# You can download 'speech.txt' file from https://github.com/akmalatnamal/davlab1/blob/main/speech.txt
file_name = "speech.txt"
words = []
special_char = ['~', ':', "'", '+', '[', '\\', '@', '^', '{', '%', '(', '-', '"', '*', '|', ',', '&', '<', '`', '}', '.', '_', '=', ']', '!', '>', ';', '?', '#', '$', ')', '/']
with open('speech.txt', 'r') as file:
  for line in file:
    words_list = line.split()
    for w in words_list:
      for i in special_char:
        if i in w:
          w = w.replace(i,'')
      if w.isalpha():
        words.append(w)
  print(len(words))
  print(words)

# Your code starts here


# Your code ends here
```

```python
# Read 'stopwords.csv' file as pandas dataframe and print its head()
# You can download 'stopwords.csv' from https://github.com/akmalatnamal/davlab1/blob/main/stopwords.csv
# Your code starts here
df = pd.read_csv('stopwords.csv')

# Your code ends here
```

```python
# Selecting stopwords that belong to English language
# Your code starts here
stopwords = df[(df["Language"] == 'English')].Words.to_list()
print(len(stopwords))
print(stopwords)

# Your code ends here
```

```python
# Removing stopwords from the word list extracted from the play

# Your code starts here
revised = []
List1 = []
List2 = []
for i in words:
  if i not in stopwords:
    revised.append(i)
for i in stopwords:
  i = i.capitalize()
  List1.append(i)
for i in revised:
  if i not in List1:
    List2.append(i)
print(len(List2))
# Your code ends here
```

```python
# Creating a frequency dictionary with words as keys and count of their occurrences as values 

frequency_dict = {}
for element in List2:
   if element in frequency_dict:
      frequency_dict[element] += 1
   else:
      frequency_dict[element] = 1
for key, value in frequency_dict.items():
   print(f"{key}: {value}")
# Your code starts here


# Your code ends here
```

```python
# Sort the dictionary based on highest frequncy of occurence
# Your code starts here

frequency_dict = sorted(frequency_dict.items(), key =
             lambda kv:(kv[1], kv[0]))
# frequency_dict = frequency_dict.reverse()
print(frequency_dict)
# Your code ends here
```

```python
# Separate the frequency_list into 2 lists for easy plotting
# Your code starts here

words = []
frequency = []
for i,j in frequency_dict:
  words.append(i)
  frequency.append(j)

words.reverse()
frequency.reverse()
print(words)
print(frequency)
# Your code ends here
```

```python
# Select the top 15 words to plot, please fill the blank spaces below.
# Your code starts here
x = words[0:14]
y = frequency[0:14 ]
print(x)
print(y)
# Your code ends here
```

```python
# Plotting the words on the x-axis and frequency along the y-axis
# Your code starts here
font = {'family':'serif','color':'darkred','size':15}
fig = plt.figure(figsize = (12, 5))
plt.title("Data Analysis and Visualization",fontdict = font)
plt.xlabel("Words", fontdict = font)
plt.ylabel("Freuency",fontdict = font)
plt.bar(x, y, color ='blue',width = 0.5)
plt.show()

# Your code ends here
```
