You are given with the dataset of top 50 music tracks. 
Each row represents a distinct song. The columns (in order) are:

ID: a unique ID (i.e., 1-50)

TrackName: Name of the Track

ArtistName: Name of the Artist

Genre: the genre of the track

BeatsPerMinute: The tempo of the song.

Energy: The energy of a song - the higher the value, the more energetic. song

Danceability: The higher the value, the easier it is to dance to this song.

Loudness: The higher the value, the louder the song.

Liveness: The higher the value, the more likely the song is a live recording. 

Valence: The higher the value, the more positive mood for the song.

Length: The duration of the song (in seconds).

Acousticness: The higher the value, the more acoustic the song is.

Speechiness: The higher the value, the more spoken words the song contains.

Popularity: The higher the value, the more popular the song is.

```python
# Import pyplot from matplotlib as plt and draw a scatter plot between Energy abd Valency Columns
import matplotlib.pyplot as plt
Energy = df['Energy'].tolist()
Valency = df['Valence'].tolist()
fig, ax = plt.subplots()
plt.xlabel('Energy')
plt.ylabel('Valency')
plt.title('Scatter plot between Energy and Valency')
ax.scatter(Energy, Valency, color = 'blue')
```

```python
# Using the above scatter plot perform the following
# Draw a horizental line using axhline and a vertical line using axvline on these two points mentioned below
# For horizental line choose a vertical point and draw it
# For vertical line choose a horizental point and draw it
# Your code here.

plt.figure(figsize=(7, 5))
plt.xlabel('Energy')
plt.ylabel('Valency')
plt.title('Scatter plot between Energy and Valency')
plt.axvline(x = 60, color = 'green')
plt.axhline(y = 30, color = 'red')
plt.scatter(Energy, Valency)
```

**Question:** Why we use axhline and axvline and why we give x as input in axvline and why we give y as input in axhline? 
Please explain this briefly.

**Answer:**
Axhline plots a horizontal line in the direction of y, starting from xmin to xmax, which ought to be somewhere in the range of 0.0 and 1.0, where 0.0 is the plot's extreme left and 1.0 is the plot's extreme right. Since axhline defines a horizontal line at area y, we utilise it as an input.

Axvline plots a vertical line in the direction of x, starting from ymin to ymax, which ought to be somewhere in the range of 0.0 and 1.0, where 0.0 is the plot's extreme bottom and 1.0 is the plot's extreme top. Since axvline defines a vertical line at area x, we utilise it as an input.



```python
# Set the figsize for your plot to (12, 16)
# Filter your DataFrame based on the Dancccceability > 60 and store it in dfd
# Draw a horizental bar plot between ArtistName and Danceability
# Show the danceability score number on the bar plot

# Your code here

dfd = df[df.Danceability > 60]
ArtistName = dfd['ArtistName'].tolist()
Danceability = dfd['Danceability'].tolist()
fig, ax = plt.subplots(figsize=(14, 14))
ax.barh(ArtistName, Danceability)
ax.set_xlabel('Danceability')
ax.set_ylabel('ArtistName')
ax.set_title('Bar plot Horizontally')
plt.show()
```

```python
# Set the figsize for your plot to (10, 6)
# Draw a line plot between Unnamed: 0 and Speechiness
# Draw another line plot between Unnamed: 0 and Acousticness
# Give a proper title and axis labels to the plot and show legends
# Annotate the points on the graph with whichever style and text you want where:
#   1. Speechiness have the value of 46
#   2. Acousticness have the value of 8

# Your code here

figure, axis = plt.subplots(figsize = (14, 6))
Unnamed = df['Unnamed: 0'].tolist()
Speechiness = df['Speechiness'].tolist()
Acousticness = df['Acousticness'].tolist()
axis.plot(Unnamed, Speechiness, label = 'Speechiness Values')
axis.plot(Unnamed, Acousticness, label = 'Acousticness Values')
axis.set_xlabel('Unnamed: 0')
axis.set_ylabel('Speechiness & Acousticness')
axis.set_title('Speechiness & Acousticness Plot')
axis.annotate('Speechiness value of 46', xy = (3, 46), xycoords='data', xytext=(5, 25), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=1.7'))
axis.annotate('Acousticness value of 8', xy = (9, 8), xycoords='data', xytext=(5, 25), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=1.7'))
axis.legend()
plt.show()
```

```python
# Sort the popularity column in descending order and plot the horizental bar plot of top 10 artists/singers based on popularity
 
DF_New = df.sort_values(by='Popularity', ascending=False)
DF_New = DF_New.head(10)
artists = DF_New['ArtistName'].tolist()
popularity = DF_New['Popularity'].tolist()
plt.figure(figsize=(7, 5))
plt.barh(artists, popularity, color = 'blue')
plt.ylabel('Artists/Singers')
plt.xlabel('Popularity')
plt.show()
```

```python
# Count the genre wise tracks and draw a bar plot in ascending order with the count of the tracks

count = df['Genre'].to_list()
count = sorted({x:count.count(x) for x in count}.items(), key=lambda x: x[1])
figure, axis = plt.subplots(figsize = (40, 10))
axis.bar([i[0] for i in count], [i[1] for i in count])
figure.show()
```

```python
# Drop all of the categorical features inplace, including Unnamed: 0
# Filter the dataframe where BeatsPerMinute > 150 and draw a heatmap in matplotlib without using seaborn

df.drop(["Unnamed: 0","TrackName", "Genre","ArtistName"], axis = 1, inplace = True, errors='ignore')
BeatsPerMinute = df[df.BeatsPerMinute > 150]
figure, axis = plt.subplots(figsize = (20, 10))
axis.imshow(BeatsPerMinute, cmap='hot')
```

```python
# Create a DataFrame sp where Speechiness > 20 and draw a bar plot between TrackName and Speechiness

from IPython.core.pylabtools import figsize
sp = df[df.Speechiness > 20]
Track_Name = sp['TrackName'].tolist()
Speechy = sp['Speechiness'].tolist()
figure, axis = plt.subplots(figsize = (24, 5))
axis.bar(Track_Name, Speechy, label = 'Speechiness Bar')
axis.set_xlabel('TrackName')
axis.set_ylabel('Speechiness')
axis.set_title('Bar plot between TrackName and Speechiness')
plt.show()
```

```python
# Draw a horizental bar plot between ArtistName and (Energy and Length) - These will be two plots together

ArtistName = df['ArtistName'].tolist()
Energy = df['Energy'].tolist()
Length = df['Length'].tolist()

figure, axis = plt.subplots(1, 2, figsize = (18,10))

axis[0].barh(ArtistName, Energy, label = 'Energy Values')
axis[0].set_xlabel('Energy')
axis[0].set_ylabel('ArtistName')
axis[0].set_title('ArtistName & Energy Plot')
axis[0].legend()

axis[1].barh(ArtistName, Length, label = 'Length Values')
axis[1].set_xlabel('Length')
axis[1].set_ylabel('ArtistName')
axis[1].set_title('ArtistName & Length Plot')
axis[1].legend()
plt.show()
```
